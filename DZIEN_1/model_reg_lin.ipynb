```r
library(ggplot2)
library(tidyverse)
```

```r
dataFrame<-read.csv('income.data_.csv')
head(dataFrame)
```

```r
x<-dataFrame$income
y<-dataFrame$happiness
plot(x,y)
```

```r
model<-lm(y~x)
model
```

```r
model_summary<-summary(model)
intercept_value<-model_summary$coefficients[1,1]
intercept_value
```
